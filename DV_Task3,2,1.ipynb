```python
# ==========================
# Bivariate Analysis (Fixed Layout) - Retail Sales Dataset
# ==========================
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.graphics.mosaicplot import mosaic
import joypy

# 1. Read dataset
file_name = "retail_sales_dataset.csv"
df = pd.read_csv(file_name)

# Pick important variables
cat1, cat2 = "Gender", "Product Category"    # categorical vs categorical
cont1, cont2 = "Quantity", "Total Amount"    # continuous vs continuous
cat3, cont3 = "Product Category", "Total Amount"  # categorical vs continuous

# ==========================
# 1. Categorical vs Categorical
# ==========================
plt.figure(figsize=(20,12))

# Stacked Bar Chart
plt.subplot(2,2,1)
ct = pd.crosstab(df[cat1], df[cat2])
ct.plot(kind="bar", stacked=True, ax=plt.gca(), colormap="viridis")
plt.title(f"Stacked Bar Chart: {cat1} vs {cat2}")
plt.xticks(rotation=30, ha="right")

# Grouped Bar Chart
plt.subplot(2,2,2)
sns.countplot(data=df, x=cat1, hue=cat2, palette="Set2")
plt.title(f"Grouped Bar Chart: {cat1} vs {cat2}")
plt.xticks(rotation=30, ha="right")

# Segmented Bar Chart
plt.subplot(2,2,3)
(ct.T / ct.T.sum()).T.plot(kind="bar", stacked=True, ax=plt.gca(), colormap="coolwarm")
plt.title(f"Segmented Bar Chart: {cat1} vs {cat2}")
plt.xticks(rotation=30, ha="right")

# Mosaic Plot
plt.subplot(2,2,4)
mosaic(df, [cat1, cat2])
plt.title(f"Mosaic Plot: {cat1} vs {cat2}")

plt.tight_layout(rect=[0,0,1,0.96])
plt.show()


# ==========================
# 2. Continuous vs Continuous
# ==========================
plt.figure(figsize=(14,6))

# Scatter Plot
plt.subplot(1,2,1)
sns.scatterplot(data=df, x=cont1, y=cont2, alpha=0.5)
plt.title(f"Scatter Plot: {cont1} vs {cont2}")

# Line Plot
plt.subplot(1,2,2)
sns.lineplot(x=df.index, y=df[cont1], label=cont1)
sns.lineplot(x=df.index, y=df[cont2], label=cont2)
plt.title(f"Line Plot: {cont1} & {cont2}")
plt.legend()

plt.tight_layout(rect=[0,0,1,0.96])
plt.show()


# ==========================
# 3. Categorical vs Continuous
# ==========================
plt.figure(figsize=(20,14))

# Bar Chart (Mean)
plt.subplot(2,3,1)
df.groupby(cat3)[cont3].mean().plot(kind="bar", color="skyblue")
plt.title(f"Bar Chart (Mean): {cat3} vs {cont3}")
plt.xticks(rotation=30, ha="right")

# Grouped KDE
plt.subplot(2,3,2)
for cat in df[cat3].unique():
    sns.kdeplot(df[df[cat3]==cat][cont3], label=cat, fill=True)
plt.title(f"Grouped KDE Plot: {cat3} vs {cont3}")
plt.legend()

# Box Plot
plt.subplot(2,3,3)
sns.boxplot(x=cat3, y=cont3, data=df, palette="Set3")
plt.title(f"Box Plot: {cat3} vs {cont3}")
plt.xticks(rotation=30, ha="right")

# Violin Plot
plt.subplot(2,3,4)
sns.violinplot(x=cat3, y=cont3, data=df, palette="muted")
plt.title(f"Violin Plot: {cat3} vs {cont3}")
plt.xticks(rotation=30, ha="right")

# Beeswarm Plot
plt.subplot(2,3,5)
sns.swarmplot(x=cat3, y=cont3, data=df, size=2)
plt.title(f"Beeswarm Plot: {cat3} vs {cont3}")
plt.xticks(rotation=30, ha="right")

# Ridgeline Plot
plt.subplot(2,3,6)
joypy.joyplot(df, by=cat3, column=cont3, figsize=(10,6), colormap=plt.cm.Set2)
plt.title(f"Ridgeline Plot: {cat3} vs {cont3}")

plt.tight_layout(rect=[0,0,1,0.96])
plt.show()
```

```python
# ==========================
# Univariate Analysis (One Variable Example) - Retail Sales Dataset
# ==========================

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Read dataset
file_name = "retail_sales_dataset.csv"
df = pd.read_csv(file_name)

# 2. Choose variables
cat_var = 'Product Category'   # categorical
cont_var = 'Total Amount'      # continuous

# ==========================
# Categorical Variable: Product Category
# ==========================
plt.figure(figsize=(14,6))

# Bar Plot
plt.subplot(1,2,1)
sns.countplot(data=df, x=cat_var, order=df[cat_var].value_counts().index, palette="viridis")
plt.title(f'Bar Plot of {cat_var}')
plt.xticks(rotation=45)

# Pie Chart
plt.subplot(1,2,2)
df[cat_var].value_counts().plot.pie(
    autopct='%1.1f%%',
    colors=sns.color_palette("viridis", len(df[cat_var].unique()))
)
plt.title(f'Pie Chart of {cat_var}')
plt.ylabel("")

plt.tight_layout()
plt.show()

# ==========================
# Continuous Variable: Total Amount
# ==========================
plt.figure(figsize=(18,12))

# Histogram
plt.subplot(3,3,1)
sns.histplot(df[cont_var], kde=False, color="skyblue")
plt.title(f'Histogram of {cont_var}')

# Density Plot
plt.subplot(3,3,2)
sns.kdeplot(df[cont_var], shade=True, color="green")
plt.title(f'Density Plot of {cont_var}')

# Rug Plot
plt.subplot(3,3,3)
sns.kdeplot(df[cont_var], shade=True, color="orange")
sns.rugplot(df[cont_var], color="black")
plt.title(f'Rug Plot of {cont_var}')

# Strip Plot
plt.subplot(3,3,4)
sns.stripplot(x=df[cont_var], color="purple")
plt.title(f'Strip Plot of {cont_var}')

# Swarm Plot
plt.subplot(3,3,5)
try:
    sns.swarmplot(x=df[cont_var], color="red", size=2)
except Exception:
    plt.text(0.5, 0.5, "Swarm plot too large", ha="center", va="center")
plt.title(f'Swarm Plot of {cont_var}')

# Scatter Plot (index vs variable)
plt.subplot(3,3,6)
plt.scatter(df.index, df[cont_var], alpha=0.5, color="blue")
plt.title(f'Scatter Plot of {cont_var}')

# Line Plot (index vs variable)
plt.subplot(3,3,7)
plt.plot(df.index, df[cont_var], alpha=0.7, color="brown")
plt.title(f'Line Plot of {cont_var}')

plt.tight_layout()
plt.show()
```
